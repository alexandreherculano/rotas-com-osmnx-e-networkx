```python
#importar bibliotecas
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, LineString
import plotly_express as px
import networkx as nx
import osmnx as ox
ox.config(use_cache=True, log_console=True)
```

```python
#função para gerar gráfico de malha viária
def create_graph(loc, dist, transport_mode, loc_type="address"):
    if loc_type == 'address':
        G = ox.graph_from_address(loc, dist=dist, network_type=transport_mode)
    elif loc_type == 'points':
        G = ox.graph_from_point(loc, dist=dist, network_type=transport_mode )
    return G
```

```python
#plotar gráfico de malha viária
G = create_graph('Natal - RN - Brazil', 2000, 'drive')
ox.plot_graph(G, bgcolor='w', figsize=(12, 15), node_color='b', node_size = 10)
```

```python
G = ox.add_edge_speeds(G)
G = ox.add_edge_travel_times(G)

#definir pontos de partida e chegada
start = (-5.819615, -35.225280)
end = (-5.812699, -35.213006)
start_node = ox.get_nearest_node(G, start)
end_node = ox.get_nearest_node(G, end)

#calcular caminho mais curto
route = nx.shortest_path(G, start_node, end_node, weight='travel_time')

#plotar rota
ox.plot_graph_route(G, route, route_linewidth=6, node_size=0, bgcolor='w')
```
